```python
def time_interval(ff):
    def inner(*args):
        if time.localtime().tm_hour in range(9, 18) or time.localtime().tm_hour == 18 and time.localtime().tm_min == 0:
            print(time.strftime("%Y %b %a, %H:%M"))
            return ff(*args)
        else:
            print("Sorry, it's too late.")
    return inner


@time_interval
def f(a):
    print(f'Welcome {a}!')

    
f('Tatev')
```
